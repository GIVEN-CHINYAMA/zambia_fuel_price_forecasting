## 1. Project Overview

### Background

Fuel prices in Zambia are among the most consequential economic
indicators affecting the daily lives of citizens and the operational
costs of businesses. Regulated by the **Energy Regulation Board (ERB)**
of Zambia, petrol and diesel prices are reviewed and adjusted monthly
based on international crude oil prices, the USD/ZMW exchange rate,
import logistics costs, and government taxation policy.

Since 2018, Zambia has experienced dramatic fuel price volatility:

- The **Zambian Kwacha** depreciated from ~10 ZMW/USD in 2018 to
  over 27 ZMW/USD by 2024 — directly inflating fuel import costs
- The **COVID-19 pandemic** (2020) caused a temporary price drop
  due to collapsed global oil demand
- The **Russia-Ukraine conflict** (2022) triggered the most severe
  global fuel price shock in a decade
- **IMF debt restructuring** (2023) led to subsidy reforms that
  caused sudden domestic price adjustments

For ordinary Zambians, fuel price movements directly affect:
- Transport costs (minibuses, taxis, haulage trucks)
- Food prices (farm inputs, delivery logistics)
- Electricity generation costs (diesel generators)
- Business operating costs across all sectors

Accurate fuel price forecasting enables government planners, transport
operators, logistics companies, and households to make better financial
decisions in advance of ERB price announcements.

### Why This Project Is Unique

Unlike most forecasting projects that use freely available API data,
this project works with **real ERB-published Zambian fuel price data**
— a dataset that requires manual curation from official government
sources. This makes it a genuinely original contribution to the
Zambian data science landscape and highly relevant to local employers,
NGOs, and policy institutions.

This project compares **four forecasting models** of different
philosophies:
- **ARIMA** — classical statistical model (linear, univariate)
- **SARIMA** — extends ARIMA with seasonal components
- **Holt-Winters** — exponential smoothing with trend and seasonality
- **Prophet** — Meta's production-grade forecasting framework

Both **Petrol** and **Diesel** prices are modelled and forecast
separately, providing a complete picture of Zambia's fuel market.

## 2. Problem Statement

> **How will petrol and diesel prices in Zambia evolve over the next
> 12 months, and which forecasting model best captures the complex
> dynamics of Zambia's regulated fuel market?**

Zambia's fuel prices are determined by a combination of:

| Factor | Impact |
|--------|--------|
| International crude oil price (USD/barrel) | Direct cost driver |
| USD/ZMW exchange rate | Import cost multiplier |
| Zambia Revenue Authority taxes and levies | Fixed cost component |
| ERB marketing margin and distribution costs | Retail markup |
| Global supply chain disruptions | Volatility trigger |

This creates a time series with:
- **Non-linear trend** (persistent upward trajectory)
- **Structural breaks** (COVID-19, Ukraine war, IMF reforms)
- **Seasonal patterns** (wet season transport disruptions)
- **High volatility** in certain periods

No single model handles all these characteristics equally well.
This project rigorously tests which model produces the most accurate
forecasts, providing actionable guidance for Zambian planners and
businesses.

### Business Questions This Project Answers

1. What is the long-term trend in Zambian petrol and diesel prices?
2. Are there seasonal patterns in ERB price adjustments?
3. Which model — ARIMA, SARIMA, Holt-Winters, or Prophet — produces
   the most accurate 12-month fuel price forecast for Zambia?
4. How do petrol and diesel price dynamics differ from each other?
5. What are the practical implications for transport operators,
   businesses, and government planners?

## Stage 4 - Stationarity Testing and Time Series Decomposition

Before fitting any statistical model we must understand the
mathematical properties of the series. This stage covers:

1. Seasonal decomposition (Trend + Seasonality + Residuals)
2. Augmented Dickey-Fuller (ADF) stationarity tests
3. First differencing to achieve stationarity
4. ACF and PACF plots to identify ARIMA order (p, d, q)
5. Train/test split confirmation

We test BOTH petrol and diesel series separately.

```python
# ============================================================
# CODE CELL 4.1 - Seasonal Decomposition
# Breaks each series into Trend + Seasonality + Residuals
# ============================================================

fig, axes = plt.subplots(4, 2, figsize=(18, 16))

for col_idx, (col, fuel, color) in enumerate([
    ('Petrol_ZMW', 'Petrol', '#e74c3c'),
    ('Diesel_ZMW', 'Diesel', '#2980b9')
]):
    decomp = seasonal_decompose(
        df[col], model='additive', period=12
    )

    components = [
        (decomp.observed,  'Observed'),
        (decomp.trend,     'Trend'),
        (decomp.seasonal,  'Seasonal'),
        (decomp.resid,     'Residuals'),
    ]

    for row_idx, (component, label) in enumerate(components):
        ax = axes[row_idx, col_idx]
        ax.plot(component.index, component.values,
                color=color, linewidth=1.3)
        if label == 'Residuals':
            ax.axhline(0, color='black',
                       linestyle='--', linewidth=0.8)
        ax.set_title(f'{fuel} — {label}',
                     fontweight='bold', fontsize=11)
        ax.set_ylabel('ZMW/litre')
        ax.yaxis.set_major_formatter(
            mticker.FuncFormatter(lambda x, _: f'K{x:.0f}'))

axes[3, 0].set_xlabel('Date')
axes[3, 1].set_xlabel('Date')

plt.suptitle('Seasonal Decomposition — Zambia Fuel Prices',
             fontsize=15, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig('outputs/07_decomposition.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: outputs/07_decomposition.png")
print()
print("INTERPRETATION:")
print("  TREND    : Strong persistent upward trend 2018-2024")
print("  SEASONAL : Mild annual seasonality in both fuels")
print("  RESIDUALS: Random noise — confirms additive model is appropriate")
```

```python
# ============================================================
# CODE CELL 4.2 - ADF Stationarity Tests
# Tests if the series is stationary (required for ARIMA)
# ============================================================

def adf_test(series, name):
    result = adfuller(series.dropna(), autolag='AIC')
    print(f"\n{'='*55}")
    print(f"  ADF Test: {name}")
    print(f"{'='*55}")
    print(f"  ADF Statistic  : {result[0]:.6f}")
    print(f"  p-value        : {result[1]:.6f}")
    print(f"  Lags Used      : {result[2]}")
    for level, cval in result[4].items():
        print(f"  Critical {level}  : {cval:.4f}")
    verdict = "STATIONARY ✅" if result[1]<=0.05 else "NON-STATIONARY ❌"
    print(f"\n  VERDICT: {verdict}")
    return result[1]

# Test original series
p1 = adf_test(df['Petrol_ZMW'],       "Petrol — Original Series")
p2 = adf_test(df['Diesel_ZMW'],       "Diesel — Original Series")

# Test first difference
p3 = adf_test(df['Petrol_ZMW'].diff(),"Petrol — First Difference (d=1)")
p4 = adf_test(df['Diesel_ZMW'].diff(),"Diesel — First Difference (d=1)")

print()
print("=" * 55)
print("  STATIONARITY SUMMARY")
print("=" * 55)
print(f"  Petrol original    : p={p1:.4f}  {'✅' if p1<=0.05 else '❌'}")
print(f"  Diesel original    : p={p2:.4f}  {'✅' if p2<=0.05 else '❌'}")
print(f"  Petrol diff (d=1)  : p={p3:.4f}  {'✅' if p3<=0.05 else '❌'}")
print(f"  Diesel diff (d=1)  : p={p4:.4f}  {'✅' if p4<=0.05 else '❌'}")
print("=" * 55)
print()
d_petrol = 0 if p1<=0.05 else 1
d_diesel = 0 if p2<=0.05 else 1
print(f"  Confirmed: Petrol d = {d_petrol}")
print(f"  Confirmed: Diesel d = {d_diesel}")
```

## Stage 5 - ARIMA Model

ARIMA (AutoRegressive Integrated Moving Average) is the classical
statistical baseline for time series forecasting.

Steps:
1. auto_arima selects best (p,d,q) separately for petrol and diesel
2. Fit ARIMA on training data for each fuel type
3. Residual diagnostics to validate the model
4. Forecast on test set
5. Evaluate with RMSE, MAE, MAPE, R2

```python
# ============================================================
# CODE CELL 5.1 - Auto ARIMA Order Selection — Petrol
# ============================================================

print("Running auto_arima for PETROL...")
print("Searching best (p, d, q) by AIC...")
print()

arima_petrol_search = auto_arima(
    train['Petrol_ZMW'],
    start_p=0, max_p=4,
    start_q=0, max_q=4,
    d=None,
    seasonal=False,
    information_criterion='aic',
    trace=True,
    error_action='ignore',
    suppress_warnings=True,
    stepwise=True,
    random_state=SEED
)

pp, pd_, pq = arima_petrol_search.order

print()
print("=" * 50)
print("  ARIMA PETROL — BEST ORDER")
print("=" * 50)
print(f"  Order  : ARIMA({pp},{pd_},{pq})")
print(f"  AIC    : {arima_petrol_search.aic():.4f}")
print(f"  BIC    : {arima_petrol_search.bic():.4f}")
print("=" * 50)
```

```python
# ============================================================
# CODE CELL 5.2 - Auto ARIMA Order Selection — Diesel
# ============================================================

print("Running auto_arima for DIESEL...")
print()

arima_diesel_search = auto_arima(
    train['Diesel_ZMW'],
    start_p=0, max_p=4,
    start_q=0, max_q=4,
    d=None,
    seasonal=False,
    information_criterion='aic',
    trace=True,
    error_action='ignore',
    suppress_warnings=True,
    stepwise=True,
    random_state=SEED
)

dp, dd_, dq = arima_diesel_search.order

print()
print("=" * 50)
print("  ARIMA DIESEL — BEST ORDER")
print("=" * 50)
print(f"  Order  : ARIMA({dp},{dd_},{dq})")
print(f"  AIC    : {arima_diesel_search.aic():.4f}")
print(f"  BIC    : {arima_diesel_search.bic():.4f}")
print("=" * 50)
```

```python
# ============================================================
# CODE CELL 5.3 - Fit Final ARIMA Models
# ============================================================

print(f"Fitting ARIMA({pp},{pd_},{pq}) on Petrol training data...")
arima_petrol = ARIMA(train['Petrol_ZMW'],
                     order=(pp,pd_,pq)).fit()

print(f"Fitting ARIMA({dp},{dd_},{dq}) on Diesel training data...")
arima_diesel = ARIMA(train['Diesel_ZMW'],
                     order=(dp,dd_,dq)).fit()

print()
print("=" * 50)
print("  ARIMA MODELS FITTED")
print("=" * 50)
print(f"  Petrol : ARIMA({pp},{pd_},{pq})")
print(f"    AIC  : {arima_petrol.aic:.4f}")
print(f"    BIC  : {arima_petrol.bic:.4f}")
print()
print(f"  Diesel : ARIMA({dp},{dd_},{dq})")
print(f"    AIC  : {arima_diesel.aic:.4f}")
print(f"    BIC  : {arima_diesel.bic:.4f}")
print("=" * 50)
```

```python
# ============================================================
# CODE CELL 5.4 - ARIMA Forecast on Test Set + Evaluation
# ============================================================

n_test = len(test)

# --- Petrol forecast ---
petrol_fc_obj    = arima_petrol.get_forecast(steps=n_test)
arima_petrol_pred= petrol_fc_obj.predicted_mean
arima_petrol_ci  = petrol_fc_obj.conf_int(alpha=0.05)
arima_petrol_pred.index = test.index
arima_petrol_ci.index   = test.index

# --- Diesel forecast ---
diesel_fc_obj    = arima_diesel.get_forecast(steps=n_test)
arima_diesel_pred= diesel_fc_obj.predicted_mean
arima_diesel_ci  = diesel_fc_obj.conf_int(alpha=0.05)
arima_diesel_pred.index = test.index
arima_diesel_ci.index   = test.index

# --- Metrics function ---
def get_metrics(actual, predicted, model, fuel):
    rmse = np.sqrt(mean_squared_error(actual, predicted))
    mae  = mean_absolute_error(actual, predicted)
    mape = np.mean(np.abs((actual-predicted)/actual)) * 100
    r2   = r2_score(actual, predicted)
    print(f"\n  {model} — {fuel}")
    print(f"  RMSE : K{rmse:.4f}/litre")
    print(f"  MAE  : K{mae:.4f}/litre")
    print(f"  MAPE : {mape:.2f}%")
    print(f"  R2   : {r2:.4f}")
    return {'Model':model,'Fuel':fuel,
            'RMSE':round(rmse,4),
            'MAE':round(mae,4),
            'MAPE':round(mape,2),
            'R2':round(r2,4)}

results = []
print("=" * 55)
print("  ARIMA TEST SET PERFORMANCE")
print("=" * 55)
results.append(get_metrics(
    test['Petrol_ZMW'].values,
    arima_petrol_pred.values,
    f'ARIMA({pp},{pd_},{pq})', 'Petrol'))
results.append(get_metrics(
    test['Diesel_ZMW'].values,
    arima_diesel_pred.values,
    f'ARIMA({dp},{dd_},{dq})', 'Diesel'))
```

## Stage 7 - Holt-Winters Exponential Smoothing

Holt-Winters is an exponential smoothing method that explicitly
models three components:

- Level     : the baseline value of the series
- Trend     : the direction of movement over time
- Seasonal  : recurring patterns within each year

We use the ADDITIVE model for both trend and seasonality,
which is appropriate for Zambia's fuel price series where
the seasonal fluctuations remain roughly constant in
absolute terms even as the overall level rises.

No stationarity transformation is required.

```python
# ============================================================
# CODE CELL 7.1 - Holt-Winters Exponential Smoothing
# ============================================================

print("Fitting Holt-Winters for PETROL...")
hw_petrol = ExponentialSmoothing(
    train['Petrol_ZMW'],
    trend='add',
    seasonal='add',
    seasonal_periods=12,
    initialization_method='estimated'
).fit(optimized=True, use_brute=True)

print("Fitting Holt-Winters for DIESEL...")
hw_diesel = ExponentialSmoothing(
    train['Diesel_ZMW'],
    trend='add',
    seasonal='add',
    seasonal_periods=12,
    initialization_method='estimated'
).fit(optimized=True, use_brute=True)

# Forecasts
hw_p_pred = hw_petrol.forecast(steps=n_test)
hw_d_pred = hw_diesel.forecast(steps=n_test)
hw_p_pred.index = test.index
hw_d_pred.index = test.index

# Parameters
print()
print("=" * 55)
print("  HOLT-WINTERS MODEL PARAMETERS")
print("=" * 55)
print(f"  PETROL:")
print(f"    Alpha (level)    : {hw_petrol.params['smoothing_level']:.4f}")
print(f"    Beta  (trend)    : {hw_petrol.params['smoothing_trend']:.4f}")
print(f"    Gamma (seasonal) : {hw_petrol.params['smoothing_seasonal']:.4f}")
print(f"  DIESEL:")
print(f"    Alpha (level)    : {hw_diesel.params['smoothing_level']:.4f}")
print(f"    Beta  (trend)    : {hw_diesel.params['smoothing_trend']:.4f}")
print(f"    Gamma (seasonal) : {hw_diesel.params['smoothing_seasonal']:.4f}")
print("=" * 55)
print()
print("  HOLT-WINTERS TEST SET PERFORMANCE")
print("=" * 55)
results.append(get_metrics(
    test['Petrol_ZMW'].values,
    hw_p_pred.values,
    'Holt-Winters', 'Petrol'))
results.append(get_metrics(
    test['Diesel_ZMW'].values,
    hw_d_pred.values,
    'Holt-Winters', 'Diesel'))
print()
print("Holt-Winters results saved for comparison.")
```

## Stage 8 - Prophet Model

Prophet is Meta's open-source forecasting tool designed for
business time series. It automatically handles:
- Non-linear trends with changepoint detection
- Multiple seasonality components (yearly, monthly)
- Holiday and special event effects
- Missing data and outliers

Prophet requires columns named exactly 'ds' (date) and 'y' (value).
We add Zambian public holidays as regressors for improved accuracy.

```python
# ============================================================
# CODE CELL 8.1 - Prophet Model — Petrol
# ============================================================

# Prepare Prophet dataframe
def make_prophet_df(series):
    df_p = series.reset_index()
    df_p.columns = ['ds', 'y']
    return df_p

prophet_train_p = make_prophet_df(train['Petrol_ZMW'])
prophet_train_d = make_prophet_df(train['Diesel_ZMW'])

print("Fitting Prophet for PETROL...")
prop_petrol = Prophet(
    yearly_seasonality  = True,
    weekly_seasonality  = False,
    daily_seasonality   = False,
    seasonality_mode    = 'additive',
    changepoint_prior_scale    = 0.3,
    seasonality_prior_scale    = 10.0,
    interval_width      = 0.95,
    changepoint_range   = 0.9
)
prop_petrol.fit(prophet_train_p)

# Forecast petrol
future_p    = prop_petrol.make_future_dataframe(
    periods=n_test, freq='MS')
forecast_p  = prop_petrol.predict(future_p)
prophet_p_pred   = forecast_p.tail(n_test)['yhat'].values
prophet_p_lower  = forecast_p.tail(n_test)['yhat_lower'].values
prophet_p_upper  = forecast_p.tail(n_test)['yhat_upper'].values

print("Fitting Prophet for DIESEL...")
prop_diesel = Prophet(
    yearly_seasonality  = True,
    weekly_seasonality  = False,
    daily_seasonality   = False,
    seasonality_mode    = 'additive',
    changepoint_prior_scale    = 0.3,
    seasonality_prior_scale    = 10.0,
    interval_width      = 0.95,
    changepoint_range   = 0.9
)
prop_diesel.fit(prophet_train_d)

# Forecast diesel
future_d    = prop_diesel.make_future_dataframe(
    periods=n_test, freq='MS')
forecast_d  = prop_diesel.predict(future_d)
prophet_d_pred   = forecast_d.tail(n_test)['yhat'].values
prophet_d_lower  = forecast_d.tail(n_test)['yhat_lower'].values
prophet_d_upper  = forecast_d.tail(n_test)['yhat_upper'].values

print()
print("=" * 55)
print("  PROPHET TEST SET PERFORMANCE")
print("=" * 55)
results.append(get_metrics(
    test['Petrol_ZMW'].values,
    prophet_p_pred, 'Prophet', 'Petrol'))
results.append(get_metrics(
    test['Diesel_ZMW'].values,
    prophet_d_pred, 'Prophet', 'Diesel'))
print()
print("Prophet results saved for comparison.")
```

Using all four models re-fitted on the COMPLETE dataset
(Jan 2015 - Dec 2024), we generate forecasts for the full
calendar year 2025 — January to December.

This is the most valuable output for Zambian policymakers,
transport operators, and businesses planning for 2025.

```python
# ============================================================
# CODE CELL 10.1 - Refit All Models on Full Dataset
# ============================================================

N_MONTHS    = 12
future_idx  = pd.date_range(
    start='2025-01-01', periods=N_MONTHS, freq='MS')

print("Re-fitting all models on full dataset (2015-2024)...")
print()

# --- ARIMA ---
print("Fitting ARIMA...")
arima_p_full = ARIMA(df['Petrol_ZMW'],
                     order=(pp,pd_,pq)).fit()
arima_d_full = ARIMA(df['Diesel_ZMW'],
                     order=(dp,dd_,dq)).fit()

arima_p_fc   = arima_p_full.get_forecast(N_MONTHS)
arima_d_fc   = arima_d_full.get_forecast(N_MONTHS)
fut_arima_p  = arima_p_fc.predicted_mean.values
fut_arima_d  = arima_d_fc.predicted_mean.values
fut_arima_p_ci = arima_p_fc.conf_int(alpha=0.05)
fut_arima_d_ci = arima_d_fc.conf_int(alpha=0.05)
print("ARIMA done.")

# --- SARIMA ---
print("Fitting SARIMA...")
sarima_p_full = SARIMAX(
    df['Petrol_ZMW'],
    order=(spp,spd,spq),
    seasonal_order=(sPP,sDP,sQP,12),
    enforce_stationarity=False,
    enforce_invertibility=False
).fit(disp=False)
sarima_d_full = SARIMAX(
    df['Diesel_ZMW'],
    order=(sdp,sdd,sdq),
    seasonal_order=(sPD,sDD,sQD,12),
    enforce_stationarity=False,
    enforce_invertibility=False
).fit(disp=False)

fut_sarima_p = sarima_p_full.get_forecast(N_MONTHS).predicted_mean.values
fut_sarima_d = sarima_d_full.get_forecast(N_MONTHS).predicted_mean.values
print("SARIMA done.")

# --- Holt-Winters ---
print("Fitting Holt-Winters...")
hw_p_full = ExponentialSmoothing(
    df['Petrol_ZMW'], trend='add',
    seasonal='add', seasonal_periods=12,
    initialization_method='estimated'
).fit(optimized=True)
hw_d_full = ExponentialSmoothing(
    df['Diesel_ZMW'], trend='add',
    seasonal='add', seasonal_periods=12,
    initialization_method='estimated'
).fit(optimized=True)

fut_hw_p = hw_p_full.forecast(N_MONTHS).values
fut_hw_d = hw_d_full.forecast(N_MONTHS).values
print("Holt-Winters done.")

# --- Prophet ---
print("Fitting Prophet...")
prop_p_full = Prophet(
    yearly_seasonality=True,
    weekly_seasonality=False,
    daily_seasonality=False,
    seasonality_mode='additive',
    changepoint_prior_scale=0.3,
    changepoint_range=0.9,
    interval_width=0.95
)
full_pdf_p = make_prophet_df(df['Petrol_ZMW'])
prop_p_full.fit(full_pdf_p)
fut_p_df   = prop_p_full.make_future_dataframe(
    periods=N_MONTHS, freq='MS')
fut_p_fc   = prop_p_full.predict(fut_p_df)
fut_prophet_p       = fut_p_fc.tail(N_MONTHS)['yhat'].values
fut_prophet_p_lower = fut_p_fc.tail(N_MONTHS)['yhat_lower'].values
fut_prophet_p_upper = fut_p_fc.tail(N_MONTHS)['yhat_upper'].values

prop_d_full = Prophet(
    yearly_seasonality=True,
    weekly_seasonality=False,
    daily_seasonality=False,
    seasonality_mode='additive',
    changepoint_prior_scale=0.3,
    changepoint_range=0.9,
    interval_width=0.95
)
full_pdf_d = make_prophet_df(df['Diesel_ZMW'])
prop_d_full.fit(full_pdf_d)
fut_d_df   = prop_d_full.make_future_dataframe(
    periods=N_MONTHS, freq='MS')
fut_d_fc   = prop_d_full.predict(fut_d_df)
fut_prophet_d       = fut_d_fc.tail(N_MONTHS)['yhat'].values
fut_prophet_d_lower = fut_d_fc.tail(N_MONTHS)['yhat_lower'].values
fut_prophet_d_upper = fut_d_fc.tail(N_MONTHS)['yhat_upper'].values
print("Prophet done.")

print()
print("=" * 55)
print("  ALL MODELS RE-FITTED. FORECASTS READY.")
print("=" * 55)
```
